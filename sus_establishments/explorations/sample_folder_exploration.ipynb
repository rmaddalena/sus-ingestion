```python
import os

def analisar_pasta(caminho):
    total_files = 0
    total_size = 0
    
    # Verifica se o caminho existe
    if not os.path.exists(caminho):
        return "Erro: O caminho especificado não existe."

    # Percorre a pasta (incluindo subpastas, se houver)
    for root, dirs, files in os.walk(caminho):
        for file in files:
            file_path = os.path.join(root, file)
            # Ignora links simbólicos para evitar erros
            if not os.path.islink(file_path):
                total_files += 1
                total_size += os.path.getsize(file_path)

    # Conversão de bytes para MB/KB para facilitar a leitura
    size_kb = total_size / 1024
    size_mb = size_kb / 1024

    print(f"--- Análise de Pasta ---")
    print(f"Caminho: {caminho}")
    print(f"Total de arquivos: {total_files}")
    print(f"Tamanho total: {total_size} bytes")
    print(f"Tamanho aproximado: {size_mb:.2f} MB ({size_kb:.2f} KB)")

# Caminho solicitado
file_path = "/Workspace/Users/rafaelmaddalena@gmail.com/raw/METADATA/CNV"

analisar_pasta(file_path)
```
