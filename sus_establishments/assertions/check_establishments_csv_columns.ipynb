```python
import os
import pandas as pd

file_path = '/Workspace/Users/rafaelmaddalena@gmail.com/raw/STSP'

# Lista todos os arquivos .csv na pasta informada em file_path
files = [f for f in os.listdir(file_path) if f.lower().endswith('.csv')]

if not files:
    print(f"Nenhum arquivo CSV encontrado em: {file_path}")
else:
    print(f"Analisando {len(files)} arquivos...\n")
    
    # Pegamos o primeiro arquivo como referÃªncia (baseline)
    primeiro_arquivo = files[0]
    df_ref = pd.read_csv(os.path.join(file_path, primeiro_arquivo), sep=';', nrows=0)
    colunas_referencia = list(df_ref.columns)
    
    mismatches = []

    for file in files:
        # Lemos apenas a linha 0 para ganhar performance
        current_df = pd.read_csv(os.path.join(file_path, file), sep=';', nrows=0)
        colunas_atuais = list(current_df.columns)
        
        if colunas_atuais != colunas_referencia:
            mismatches.append({
                "arquivo": file,
                "colunas_faltantes": list(set(colunas_referencia) - set(colunas_atuais)),
                "colunas_extras": list(set(colunas_atuais) - set(colunas_referencia))
            })

    # --- Resultado da VerificaÃ§Ã£o ---
    if not mismatches:
        print("Sucesso! Todos os arquivos possuem as mesmas colunas e na mesma ordem.")
        print(f"Colunas detectadas ({len(colunas_referencia)}): {colunas_referencia}")
    else:
        print("AtenÃ§Ã£o! Foram encontrados arquivos com colunas diferentes:\n")
        for m in mismatches:
            print(f"ðŸ“„ Arquivo: {m['arquivo']}")
            if m['colunas_faltantes']:
                print(f"   - Faltando: {m['colunas_faltantes']}")
            if m['colunas_extras']:
                print(f"   - Extras: {m['colunas_extras']}")
            print("-" * 30)
```
